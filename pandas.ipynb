PLOTS

```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate some random numbers
randomNumber = np.random.rand(10)
randomNumber
```

```python
#plot a line graph
style.use("ggplot")
plt.plot(randomNumber,"g", label = 'line', linewidth = 2 )
plt.xlabel("range")
plt.ylabel("number")
plt.title('plot')
plt.legend()
plt.show()
```

```python

cause = "diseases","accidents","natural calamities", "old age", 'suicide'
percent = [40,15,5,20,10]
```

```python
plt.figure(figsize=(10,10))
#set the explode
explode = [0.1,0,0,0,0]
#pie properties
plt.pie(percent, labels= cause, explode = explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.legend()
plt.title('Lead death causes')
plt.show()
```
